```python
import os, time
import pathlib
import numpy as np
import scipy.io as sio
import scipy.io.wavfile
import pandas as pd

errList = []
ftime = 0
j=0

pathList = []

for (path, dir, files) in os.walk(os.getcwd()):
    for filename in files:
        ext = os.path.splitext(filename)[-1]
        if ext == '.wav':
           # path = os.path.realpath(os.path.dirname(__file__))
            spath = path.replace("\\","/") + "/" + filename
            rate, data = sio.wavfile.read(spath)
            times = np.arange(len(data))/float(rate)

            ftime += times[-1]
            j += 1
            #mktime = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(os.path.getctime(spath)))   #만든날짜
            uptime = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(os.path.getmtime(spath)))    #수정한날
            print(filename, uptime)

            pathList.append([filename, spath, times[-1], uptime])

h = ftime / 3600

tm = ftime % 3600
m = tm / 60

s = tm % 60

print(j)
print('\n', ftime)
print(int(h), ":", int(m), ":", int(s))
print('ErrorList:', errList, '\nCount:', len(errList))

pathList = pd.DataFrame(pathList)
pathList.to_excel(excel_writer='data list.xlsx')
if len(errList) > 0:
    errList = pd.DataFrame(errList)
    errList.to_excel(excel_writer='err list.xlsx')
```
